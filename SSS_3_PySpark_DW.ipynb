```
from pyspark.sql.functions import *
from pyspark.sql.functions import lit
from pyspark.sql.types import StructType,StructField, StringType, IntegerType
spark.conf.set("spark.sql.execution.arrow.enabled","true")
```

```
Orders =spark.read.format('csv').option('header','true').option('infor schema','true').load('dbfs:/FileStore/tables/Orders.csv')
display(Orders.limit(5))
```

```
Orders.printSchema()
```

```
Users_MainData = spark.read.format('csv').option('header','true').load('dbfs:/FileStore/tables/Users.csv')
Users = Users_MainData
display(Users.limit(3))
```

```
Users.printSchema()
```

```
Returns_MainData = spark.read.format('csv').option('header','true').load('dbfs:/FileStore/tables/Returns.csv')
returns = Returns_MainData
display(returns.limit(3))
```

```
returns.printSchema()
```

```
# REMOVE DUPLICATES
Orders.dropDuplicates()
returns.dropDuplicates()
Users.dropDuplicates()
```

```
#Removing Columns
Orders = Orders.drop('Ship Date')
display(Orders.limit(3))
```

```
# Columns Flters
Orders.filter(Orders.Region == 'South')
display(Orders.limit(3))
```

```
# Column Rename
Orders = Orders.withColumnRenamed('Order Date','Order_Date')
Orders = Orders.withColumnRenamed('Order ID','OrderID')
Orders = Orders.withColumnRenamed('Unit Price','UnitPrice')
returns = returns.withColumnRenamed('Order Date','Order_Date')
returns = returns.withColumnRenamed('Order ID','OrderID')
```

```
#Add a Columns
Orders = Orders.withColumn('FinalPrice',Orders.UnitPrice - Orders.Discount)
display(Orders.limit(3))
```

```
# Column Data TYpe change
Orders = Orders.withColumn('Row ID',Orders['Row ID'].cast('int'))
Orders = Orders.withColumn('Order_Date',Orders['Order_Date'].cast('TimeStamp'))
Orders = Orders.withColumn('UnitPrice',Orders['UnitPrice'].cast('float'))
Orders = Orders.withColumn("Postal Code",Orders["Postal Code"].cast('Bigint'))
Orders = Orders.withColumn("Sales",Orders["Sales"].cast('float'))
Orders = Orders.withColumn("Quantity ordered new",Orders["Quantity ordered new"].cast('int'))
Orders = Orders.withColumn("Discount",Orders["Discount"].cast('float'))
Orders = Orders.withColumn("Profit",Orders["Profit"].cast('float'))
```

```
display(Orders.limit(3))
display(returns.limit(3))
```

```
df = Orders.join(returns,Orders.OrderID == returns.OrderID,'left')
display(df)
```
